```python
#Curve with error band
#This EXAMPLE illustrates how to draw an error band around a parametrized curve.
#A parametrized curve x(t), y(t) can directly be drawn using plot.

import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import PathPatch
from matplotlib.path import Path

N = 400
t = np.linspace(0, 2 * np.pi, N)
r = 0.5 + np.cos(t)
x, y = r * np.cos(t), r * np.sin(t)

fig, ax = plt.subplots()
ax.plot(x, y, "k")
ax.set(aspect=1)
```

```python
#Sample data (dictionary) representing the average temperature (in degrees Celsius) from 2010 to 2014
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
df
```

```python
#ver1 - Matplotlib
plt.plot(df['Year'], df['Avg_Temperature'], marker='o') #df[] means refer to a column: first one is X and the second is Y (markers also: x, s, o)
plt.title('Average temperature from 2010 to 2014') #name of DF
plt.xlabel('Year') #name on X axis
plt.ylabel('Temperature') #title on Y axis
plt.xticks(df['Year']) #year values are rounded
plt.grid(True)
plt.show() #show you the visualization
```

```python
#Ver2 - Pandas library
df.plot(x='Year', y='Avg_Temperature', kind='line') #we are using Pandas library
plt.title('Average temperature from 2010 to 2014') #name of DF
plt.xlabel('Year') #name on X axis
plt.ylabel('Temperature') #title on Y axis
plt.xticks(df['Year']) #year values are rounded
plt.grid(True)
plt.show() #show you the visualization
```

```python
#BAR CHART

plt.bar(df['Year'], df['Avg_Temperature'], color='skyblue') #color can be selected by you
plt.title('Average temperature from 2010 to 2014')
plt.xlabel('Year') #title on X axis
plt.ylabel('Temperature') #title on Y axis
```

```python
#horizontal bar chart

plt.barh(df['Year'], df['Avg_Temperature'], color='orange') #color can be selected by you
plt.title('Avg temperature from 2010 to 2014')
plt.xlabel('Temperature')
plt.ylabel('Year')
```

```python
# HISTOGRAM - presents the distribution of a single variable, frequency, represents numerical data
# with bins we said that we will have 3 ranges, and then it subdivides the ax between lowest to max score into 3 equal intervals
data2 = {'Scores': [85, 92, 78, 20, 89, 90, 82, 56, 60, 23, 90, 40, 77]}
df2 = pd.DataFrame(data2)
df2

plt.hist(df2['Scores'], bins = 3, color = 'green', edgecolor = 'black')
```

```python
#SCATTER PLOT - used to visualize the relationship between two variables
data3 = {'X': [1, 2, 3, 5, 5, 2], 'Y': [5, 8, 7, 4, 9, 5]}

df3 = pd.DataFrame(data3)
df3
plt.scatter(x = df3['X'], y = df3['Y'])
plt.title('Scatter Plot')
#plt.grid(True, which='both', linestyle='--', linewidth=0.1, color='blue') - change the color of the grid
plt.grid(True)
```

```python
# PIE CHART - represents parts of a whole

data4 = {'Category': ['A', 'B', 'C', 'D'], 'Values': [15, 30, 25, 40]}
df4 = pd.DataFrame(data4)
df4

plt.pie(df4['Values'], labels = df4['Category'], autopct = '%1.1f%%')
plt.title('PIE CHART')
plt.show()
```

```python
#BOX PLOT - distribution of a dataset and identifies outliers
plt.boxplot(df4['Values'])
plt.grid(True)
plt.show()
```

```python
#Teamwork (Start with Task 1 and proceed further as complexity will increase):
#To complete the tasks use 'transaction_dataset.csv'

#Task 1: Create a bar chart that shows the count of customers
#for each unique value in the 'Gender' column (including NaN values). - easy
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('/transaction_dataset.csv')
dataset['Gender'] = dataset['Gender'].fillna('NaN')
data5 = dataset['Gender'].value_counts(dropna=False)
plt.title('Count of Customers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.bar(data5.index, data5.values, color = 'pink')
plt.grid(True)
plt.show()
```

```python
#Task 2: Create a horizontal bar chart that shows the top 5 most frequent names
#in the DataFrame, based on the 'name' column. (First, create a grouped DataFrame (name_df),
#then filter it using iloc, and finally create the visualization.) -medium
name_df = dataset['Name'].value_counts().reset_index()
name_df.columns = ['name', 'count']
top_5_names = name_df.iloc[:5]
plt.barh(top_5_names['name'], top_5_names['count'], color='pink')
plt.xlabel('Count')
plt.ylabel('Name')
plt.title('Top 5 Most Frequent Names')
plt.show()
```

```python
#Task 3: Create a filtered DataFrame that includes Category == 'Clothing' and Gender == 'M'.
#How many rows are there in this filtered DataFrame? (3738 rows × 9 columns)
#Format the result as follows: The filtered DataFrame has XXXX rows. - hard

filtered_df = dataset[(dataset['Category'] == 'Clothing') & (dataset['Gender'] == 'M')]
#print(filtered_df)
num_rows = filtered_df.shape[0]
print(f'The filtered DataFrame has {num_rows} rows.')
```
